```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
features = pd.read_csv('fma_metadata/features.csv',low_memory=False)
features=features.iloc[3:106577].reset_index(drop=True).rename(columns={'feature':'track_id'}) #removed first three rows because reading it in normally includes weird multi-indexing formatting
features['track_id']=features['track_id'].astype(int)# because of previous odd csv formatting, ids were str and not float
features
```

```python
tracks = pd.read_csv('fma_metadata/tracks.csv',low_memory=False)
tracks.columns = ['track_id']+list(tracks.iloc[0])[1:] #fixed formatting errors
tracks = tracks.iloc[2:].reset_index(drop=True)
tracks
```

```python
tracksSubset = tracks[['track_id','favorites','listens','name','title','duration','genre_top']] #removed unessecary columns
tracksSubset.columns = ['track_id','favorites_1','favorites_2','favorites_2','listens_1','listens_2','artist','album','title','duration','genre']
tracksSubset['track_id']=tracksSubset['track_id'].astype(int) #id was string
tracksSubset
```

```python
echohonest = pd.read_csv('fma_metadata/echonest.csv',low_memory=False)
echohonest.columns = ['track_id']+list(echohonest.iloc[1])[1:]
echohonest=echohonest.iloc[3:].reset_index(drop=True)
echohonest['track_id']=echohonest['track_id'].astype(int)
echohonest
```

```python
data = echohonest.join(tracksSubset,on='track_id',rsuffix='_tracks')
data=data.join(features,on='track_id',rsuffix='_features')
data=data.loc[pd.notnull(data['zcr'])] #removed songs not present in echohonest
data=data.loc[pd.notnull(data['genre'])].reset_index(drop=True)
data
```

```python
def changeFavorites(df):
    '''
    Certain variables were incorrectly formatted when the dataframes were created, this removes
    one such incident where a column name favorites_2 was repeated. It changes that to favorites_3.'''
    dfColList = list(df.columns)
    seen=0 #flag
    for i in range(len(dfColList)):
        if dfColList[i]=='favorites_2' and seen==0: #show we have seen it for the first time
            seen=1
        elif dfColList[i]=='favorites_2': #the second time change it
            dfColList[i]='favorites_3'
    dfCopy = df.copy(deep=True)
    dfCopy.columns = dfColList #change column names 
    return dfCopy
        
```

```python
data1 = data.drop(['track_id','album_date','album_name','release','artist','artist_name','album','title','genre','artist_location','artist_longitude','artist_latitude','artist_familiarity_rank','artist_discovery_rank','artist_hotttnesss_rank','song_currency_rank','song_hotttnesss_rank'],axis=1) #removed variables with majority NaN
response = data['genre'] #separate response variable for SVM
data1= data1.astype(float) #due to previous formatting errors some numeric values are strings, change to float
data1 = changeFavorites(data1) 
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder() #encode variables so that they can be put into a model
responseEncoded = le.fit_transform(response)
responseEncoded
```

```python
data = pd.read_csv('FMADatasetTrans-cleaned.csv').drop('Unnamed: 0',axis=1)

np.random.seed(100)
X_train, X_test, y_train, y_test = train_test_split(data,responseEncoded,test_size=.4)
```

```python
from imblearn.over_sampling import SMOTE
#oversampling with smote so we can train using synthetic instances
unique,counts = np.unique(y_train,return_counts=True)
sm = SMOTE(random_state=146,m_neighbors=min(counts)-1,k_neighbors=min(counts)-1)
X_SMOTE, y_SMOTE = sm.fit_sample(X_train,y_train)
```

```python
#remove impure data points
from imblearn.under_sampling import TomekLinks
tl = TomekLinks(ratio='all')

X_tl, y_tl = tl.fit_sample(X_SMOTE,y_SMOTE)
```

# Rbf

## rbf with SMOTE

```python
svmModel = SVC(kernel='rbf',C=2.2)
```

```python
svmModel.fit(X_SMOTE,y_SMOTE)
```

```python
y_pred = svmModel.predict(X_test)
```

```python
f1=f1_score(y_true=y_test,y_pred=y_pred,average='macro')
a=accuracy_score(y_true=y_test,y_pred=y_pred)
print(f1,a)
```

```python
plt.figure(figsize=(20,20))
plt.matshow(confusion_matrix(y_test, y_pred),cmap=plt.cm.Blues) 
plt.colorbar()
```
